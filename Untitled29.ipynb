```python
cat_col = df.select_dtypes(include= "object").columns
```

```python
cat_col
```

```python
for cn in cat_col:
    print(df[cn].unique())
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.impute import SimpleImputer
```

```python
for cn in cat_col:
    ode = OrdinalEncoder(categories=[sorted(df[cn].unique())])
    df[cn] = ode.fit_transform(df[[cn]])
    
```

```python
df.head()
```

```python
min(df['price'])
```

```python
max(df['price'])
```

```python
print(max(df['price']))
print(min(df['price']))
```

```python
def convertInlac(val):
    return val/100000
```

```python
df['price'] = df['price'].map(convertInlac)
```

```python
df.head()
```

```python
X = df.drop(columns=['price'])
```

```python
y = df['price']
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=11,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lg = LinearRegression()
```

```python
lg.fit(X_train,y_train)
```

```python
y_pred = lg.predict(X_test)
```

```python
y_pred.shape
```

```python
plt.scatter(y_test,y_pred, c = 'red')
plt.xlabel('Price in INR-Lac')
plt.ylabel('Prediction Price')
plt.title('Ordinal Value Vs Predicted Value : Linear Regression')
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error , r2_score
```

```python
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
```

```python
print("mean_absolute_error : ", mae)
print("mean_squared_error : ", mse)
```

```python
print("RSME :",r2_score(y_test,y_pred)*100)
```
